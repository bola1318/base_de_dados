```python
#Bibliotecas


import pandas as pd
import requests 

from google.colab  import data_table             
data_table.enable_dataframe_formatter()      
```

```python
#informações para navegador

header= {
        'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"}
 
#url para acesso
url_link1 = "https://www.soccerstats.com/matches.asp?matchday=1&listing=1"   #https://www.soccerstats.com/matches.asp?matchday=1&listing=1"
url_link2 = "https://www.soccerstats.com/matches.asp?matchday=1&listing=2"

#juntando com requests

r1= requests.get(url_link1, headers=header)
r2= requests.get(url_link2, headers=header)

#usando a função read_html do pabndas

df1 = pd.read_html(r1.text)
df2 = pd.read_html(r2.text)

```

```python
##GP: number of matches played
##W%: percentage of matches won
##FTS: % Failed to score (no color coding)
##CS: % Clean sheet (no color coding)
##BTS: % of matches where both teams scored (no color coding)
##TG: Average total goals scored + conceded per match
##GF: goals scored per match (on this page: in green, no color coding)
##GA: goals conceded per match (on this page: in red, no color coding)
##1.5+: % of matches over 1.5 goals (scored + conceded)
##2.5+: % of matches over 2.5 goals (scored + conceded) - on this page: values above 50% displayed in blue, others displayed in red.
##3.5+: % of matches over 3.5 goals (scored + conceded)
##PPG: Points per game (no color coding)
```

```python
jogos_today1 = df1[6]
jogos_today1 = jogos_today1[['Country','2.5+','1.5+','GA','GF','TG','PPG','Unnamed: 9','Unnamed: 10','Unnamed: 11','PPG.1','TG.1','GF.1','GA.1','1.5+.1','2.5+.1']]
jogos_today1.columns = ['País','Over2,5_casa_gols','Over1,5_casa_gols','GolsSofridos_casa','GolsMarcados_casa','MediaGols_casa','PontosPorJogo_casa','casa','Hora','visitante','PontosPorJogo_visitante','MediaGols_visitante','GolsMarcados_visitante','GolsSofridos_visitante','Over1,5_visitante_gols','Over2,5_visitante_gols']
```

```python
jogos_today2 = df2[6]
jogos_today2 = jogos_today2[['BTS','W%','BTS.1','W%.1']]
jogos_today2.columns = ['BTTS_casa','%Vitorias_casa','BTTS_visitante','%Vitorias_visitante']
```

```python
jogos_today = pd.concat([jogos_today1,jogos_today2],axis=1)
jogos_today = jogos_today[['País','Hora','casa','visitante','%Vitorias_casa','%Vitorias_visitante','Over1,5_casa_gols','Over2,5_casa_gols','Over1,5_visitante_gols','Over2,5_visitante_gols','BTTS_casa','BTTS_visitante','GolsMarcados_casa','GolsSofridos_casa','GolsMarcados_visitante','GolsSofridos_visitante','MediaGols_casa','MediaGols_visitante','PontosPorJogo_casa','PontosPorJogo_visitante']]
```

```python
jogos_today
```

```python
#reorganizar a tabela
jogos_today = jogos_today.sort_values('Hora')
jogos_today['Hora'] = pd.to_datetime(jogos_today['Hora']) - pd.DateOffset(hours=4)
jogos_today['Hora'] = pd.to_datetime(jogos_today['Hora'], format='%H:%M').dt.time
jogos_today = jogos_today.dropna()

# Resetando o INDEX
jogos_today.reset_index(inplace=True, drop=True)
#jogos_today.index= jogos_today.index.set_name(['N'])
jogos_today = jogos_today.rename(index=lambda x: x +1)
jogos_today.to_excel("Jogos de Hoje.xlsx")
```

```python
df = jogos_today
df
```
